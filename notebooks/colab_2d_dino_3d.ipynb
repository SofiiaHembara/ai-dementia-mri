```python
# === Пакети (у Colab уже є все потрібне) ===
import os, json, math, numpy as np, pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
from sklearn.metrics import (
    roc_curve, auc, precision_recall_curve, average_precision_score,
    confusion_matrix, f1_score, balanced_accuracy_score, brier_score_loss
)
from sklearn.calibration import calibration_curve

# === НАЛАШТУВАННЯ ШЛЯХІВ ===
ART = Path("/content/drive/MyDrive/ai-dementia-mri/ai-dementia-mri/artifacts")
POST = ART / "poster"
POST.mkdir(parents=True, exist_ok=True)

SLICE_CSV_2D = ART / "test_preds_2d_dino.csv"     # slice/зрізи 2D
SLICE_CSV_3D = ART / "test_preds_3d.csv"           # опціонально: 3D (якщо є)
THR_JSON_2D  = ART / "thresholds.json"             # якщо є обраний поріг з val

# === ДОП. ПАРАМЕТРИ ===
AGGREGATION = "mean"   # 'mean' або 'topk'
TOPK = 3               # якщо 'topk'
FIXED_TEST_THR = 0.50  # фіксований трешхолд для візуалізацій на тесті (щоб не «підбирати» по тесту)

# === ХЕЛПЕРИ ===
def savefig(base, tight=True):
    png = POST / f"{base}.png"
    pdf = POST / f"{base}.pdf"
    if tight:
        plt.savefig(png, dpi=300, bbox_inches="tight")
        plt.savefig(pdf, bbox_inches="tight")
    else:
        plt.savefig(png, dpi=300)
        plt.savefig(pdf)
    print("Saved:", png, "and", pdf)

def load_preds_csv(path: Path):
    df = pd.read_csv(path)
    # нормалізація назв колонок
    cols = {c.lower(): c for c in df.columns}
    # y (label)
    if "y" not in cols and "label" in cols:
        df = df.rename(columns={cols["label"]: "y"})
    elif "y" not in cols:
        raise ValueError(f"Не знайшов колонку y/label у {path}")
    # p (prob)
    if "p" not in cols:
        for cand in ["prob","pred","y_prob","proba"]:
            if cand in cols:
                df = df.rename(columns={cols[cand]: "p"})
                break
    if "p" not in df.columns:
        raise ValueError(f"Не знайшов колонку p/prob у {path}")
    # sid
    if "sid" not in df.columns:
        for cand in ["subject_id","patient_id","case_id","id"]:
            if cand in df.columns:
                df = df.rename(columns={cand: "sid"})
                break
    if "sid" not in df.columns:
        raise ValueError(f"Не знайшов колонку sid/subject_id у {path}")
    # to numeric
    df["y"] = pd.to_numeric(df["y"], errors="coerce").fillna(0).astype(int)
    df["p"] = pd.to_numeric(df["p"], errors="coerce")
    return df[["sid","y","p"]].copy()

def patient_aggregate(df, how="mean", topk=3):
    """ df: колонки sid,y,p (slice-level). Повертає patient-level DataFrame з y,p """
    if how == "mean":
        agg = df.groupby("sid").agg(y=("y","mean"), p=("p","mean")).reset_index()
    elif how == "median":
        agg = df.groupby("sid").agg(y=("y","mean"), p=("p","median")).reset_index()
    elif how == "topk":
        # беремо top-k найбільших p по пацієнту
        parts = []
        for sid, g in df.groupby("sid"):
            y = 1 if g["y"].mean() > 0.5 else 0
            ps = g["p"].nlargest(topk).values
            p  = float(np.mean(ps)) if len(ps) > 0 else float(g["p"].mean())
            parts.append((sid, y, p))
        agg = pd.DataFrame(parts, columns=["sid","y","p"])
    else:
        raise ValueError("how must be in {'mean','median','topk'}")
    agg["y"] = (agg["y"] > 0.5).astype(int)
    return agg

def plot_roc(y, p, title, fname):
    fpr, tpr, _ = roc_curve(y, p)
    A = auc(fpr, tpr)
    plt.figure()
    plt.plot(fpr, tpr, label=f"AUC={A:.3f}")
    plt.plot([0,1],[0,1], linestyle="--")
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.title(title)
    plt.legend(loc="lower right")
    savefig(fname)

def plot_pr(y, p, title, fname):
    pr, rc, _ = precision_recall_curve(y, p)
    AP = average_precision_score(y, p)
    plt.figure()
    plt.plot(rc, pr, label=f"AP={AP:.3f}")
    plt.xlabel("Recall")
    plt.ylabel("Precision")
    plt.title(title)
    plt.legend(loc="lower left")
    savefig(fname)

def plot_confmat(y, p, thr, title, fname):
    yhat = (p >= thr).astype(int)
    cm = confusion_matrix(y, yhat, labels=[0,1])
    tn, fp, fn, tp = cm.ravel()
    plt.figure()
    plt.imshow(cm, interpolation="nearest")
    plt.xticks([0,1], ["Pred 0","Pred 1"])
    plt.yticks([0,1], ["True 0","True 1"])
    for i, val in enumerate([tn,fp,fn,tp]):
        r, c = divmod(i, 2)
        plt.text(c, r, str(val), ha="center", va="center")
    plt.title(f"{title}\nF1={f1_score(y,yhat):.3f}  BA={balanced_accuracy_score(y,yhat):.3f}  thr={thr:.2f}")
    plt.colorbar()
    savefig(fname)

def plot_calibration(y, p, title, fname):
    prob_true, prob_pred = calibration_curve(y, p, n_bins=10, strategy="uniform")
    bs = brier_score_loss(y, p)
    plt.figure()
    plt.plot([0,1],[0,1], linestyle="--", label="ідеальна калібровка")
    plt.plot(prob_pred, prob_true, marker="o", label=f"модель (Brier={bs:.3f})")
    plt.xlabel("Середнє передбачене p")
    plt.ylabel("Частка позитивів")
    plt.title(title)
    plt.legend()
    savefig(fname)

def plot_f1_ba_vs_thr(y, p, title, fname):
    thrs = np.linspace(0.0, 1.0, 101)
    f1s, bas = [], []
    for t in thrs:
        yhat = (p >= t).astype(int)
        f1s.append(f1_score(y, yhat))
        bas.append(balanced_accuracy_score(y, yhat))
    plt.figure()
    plt.plot(thrs, f1s, label="F1")
    plt.plot(thrs, bas, label="Balanced Accuracy")
    plt.xlabel("Threshold")
    plt.ylabel("Score")
    plt.title(title)
    plt.legend()
    savefig(fname)

def plot_prob_dists(y, p, title, fname):
    plt.figure()
    plt.hist(p[y==0], bins=30, alpha=0.6, label="Class 0 (non_demented)")
    plt.hist(p[y==1], bins=30, alpha=0.6, label="Class 1 (dementia)")
    plt.xlabel("Predicted probability p")
    plt.ylabel("Count")
    plt.title(title)
    plt.legend()
    savefig(fname)

# === 2D: завантаження та агрегація ===
assert SLICE_CSV_2D.exists(), f"Немає {SLICE_CSV_2D}"
df2d = load_preds_csv(SLICE_CSV_2D)

# агрегація до patient-level
pat2d = patient_aggregate(df2d, how=AGGREGATION, topk=TOPK)

# трешхолд: на постері краще фіксований для ТЕСТ (щоб не «читити»),
# але якщо є val-поріг у thresholds.json — використаємо його для підпису
thr_used = FIXED_TEST_THR
if THR_JSON_2D.exists():
    try:
        j = json.loads(Path(THR_JSON_2D).read_text())
        if "thr" in j:
            thr_used = float(j["thr"])
    except Exception:
        pass

# === ГРАФІКИ 2D (patient-level) ===
y2, p2 = pat2d["y"].values, pat2d["p"].values
plot_roc(y2, p2, f"2D DINO, patient-level ({AGGREGATION}{'' if AGGREGATION!='topk' else f' k={TOPK}'})",
         "roc_patient_2d")
plot_pr(y2, p2,  f"2D DINO, patient-level PR", "pr_patient_2d")
plot_confmat(y2, p2, thr_used, "2D DINO, Confusion matrix @test",
             f"cm_patient_2d_thr{int(100*thr_used):02d}")
plot_calibration(y2, p2, "2D DINO, калібрування", "calibration_patient_2d")
plot_f1_ba_vs_thr(y2, p2, "2D DINO, F1/BA vs threshold", "f1_ba_vs_thr_2d")
plot_prob_dists(y2, p2, "2D DINO, розподіл ймовірностей", "hist_probs_2d")

# slice-level 2D: (за бажанням) ROC/PR теж можна домалювати:
y2s, p2s = df2d["y"].values, df2d["p"].values
plot_roc(y2s, p2s, "2D DINO, slice-level ROC", "roc_slice_2d")
plot_pr(y2s, p2s,  "2D DINO, slice-level PR",  "pr_slice_2d")

# === Порівняння моделей (опційно додамо 3D, якщо CSV існує) ===
labels, aucs = ["2D DINO (patient)"], []

# обчислимо AUC для 2D patient:
fpr, tpr, _ = roc_curve(y2, p2)
aucs.append(auc(fpr, tpr))

if SLICE_CSV_3D.exists():
    df3d = load_preds_csv(SLICE_CSV_3D)
    pat3d = patient_aggregate(df3d, how=AGGREGATION, topk=TOPK)
    y3, p3 = pat3d["y"].values, pat3d["p"].values
    fpr3, tpr3, _ = roc_curve(y3, p3)
    aucs.append(auc(fpr3, tpr3))
    labels.append("3D R3D (patient)")

# бар-діаграма
plt.figure()
plt.bar(range(len(aucs)), aucs)
plt.xticks(range(len(aucs)), labels, rotation=15, ha="right")
plt.ylabel("AUC (patient-level)")
plt.title("Порівняння моделей")
savefig("compare_models_patient_auc")
```

```python
import pandas as pd, os, numpy as np

def fix_index_paths(src_csv, dst_csv, prefer_col=None, new_root='/content/data'):
    df = pd.read_csv(src_csv)
    path_cols = []
    for c in ['slice_path', 'path', 'nifti_path']:
        if c in df.columns:
            path_cols.append(c)
    if prefer_col and prefer_col in df.columns:
        path_col = prefer_col
    else:
        # виберемо першу доступну
        path_col = path_cols[0] if path_cols else None
    if path_col is None:
        raise RuntimeError(f'Не знайшов колонку з шляхами у {src_csv}. Є колонки: {df.columns.tolist()}')

    # вичислимо спільний префікс (old_root)
    paths = df[path_col].dropna().astype(str).tolist()
    # відфільтруємо щось схоже на абсолютні шляхи
    paths = [p for p in paths if p.startswith('/') and len(p)>1]
    if not paths:
        raise RuntimeError('Немає валідних абсолютних шляхів для заміни.')
    old_root = os.path.commonprefix(paths)

    # commonprefix працює посимвольно — підрівняємо до межі каталогу
    old_root = old_root[:old_root.rfind('/')+1] if '/' in old_root else old_root
    print('old_root ->', old_root)
    print('new_root ->', new_root)

    df[path_col] = df[path_col].astype(str).str.replace(old_root, new_root + '/', regex=False)

    # Перейменуємо колонку для 3D (якщо потрібно) на 'nifti_path'
    if 'nifti_path' not in df.columns and 'path' in df.columns and 'slice_path' not in df.columns:
        df.rename(columns={'path':'nifti_path'}, inplace=True)

    # Перейменуємо колонку для 2D (якщо потрібно) на 'slice_path'
    if 'slice_path' not in df.columns and 'path' in df.columns:
        # це 2D випадок
        df.rename(columns={'path':'slice_path'}, inplace=True)

    df.to_csv(dst_csv, index=False)
    print('Saved:', dst_csv)

    # перевіримо існування файлів (вибірково)
    sample = df[path_col if path_col in df.columns else ('nifti_path' if 'nifti_path' in df.columns else 'slice_path')].dropna().head(5).tolist()
    exists = [os.path.exists(p) for p in sample]
    print('Sample exists check:', list(zip(sample, exists)))

# Фіксимо обидва
FIXED_2D = '/content/index_oasis1_2d_colab.csv'
FIXED_3D = '/content/index_oasis1_for3d_colab.csv'

fix_index_paths(
    src_csv=os.path.join(DRIVE_DATA_DIR, 'index_oasis1_2d.csv'),
    dst_csv=FIXED_2D,
    prefer_col=None,      # авто
    new_root='/content/data'
)

fix_index_paths(
    src_csv=os.path.join(DRIVE_DATA_DIR, 'index_oasis1_for3d.csv'),
    dst_csv=FIXED_3D,
    prefer_col=None,      # авто
    new_root='/content/data'
)
```

```python
import os, glob, pandas as pd

OLD_INDEX = "/content/ai-dementia-mri/data/index_oasis1_for3d.csv"
NEW_INDEX = "/content/ai-dementia-mri/data/index_oasis1_for3d_colab.csv"
ROOT      = "/content/data/processed/oasis1_nifti"

df = pd.read_csv(OLD_INDEX)
assert {"subject_id","label","split"}.issubset(df.columns), df.columns

def find_nifti_for_sid(sid: str):
    c1 = f"{ROOT}/{sid}_T1w.nii.gz"
    if os.path.exists(c1):
        return c1
    cands = [
        f"{ROOT}/{sid}.nii.gz",
        f"{ROOT}/{sid}/T1w.nii.gz",
        f"{ROOT}/{sid}/anat/{sid}_T1w.nii.gz",
    ]
    for c in cands:
        if os.path.exists(c):
            return c
    # Fallback: глобальний пошук
    g = glob.glob(f"{ROOT}/**/{sid}*nii*", recursive=True)
    return g[0] if g else None

df["nifti_path"] = df["subject_id"].apply(find_nifti_for_sid)

missing = df["nifti_path"].isna().sum()
print("❗Відсутніх шляхів після фіксу:", missing)
if missing:
    print("Напр.:", df[df["nifti_path"].isna()].head(5))

# прибираємо порожні шляхи
df = df.dropna(subset=["nifti_path"]).reset_index(drop=True)

# контроль, що всі файли існують
not_exist = (~df["nifti_path"].apply(os.path.exists)).sum()
print("❗Шлях не існує (має бути 0):", not_exist)

print("✅ Підсумок:", df["split"].value_counts().to_dict())
df.to_csv(NEW_INDEX, index=False)
print("💾 Збережено:", NEW_INDEX, "| рядків:", len(df))
```

```python
import os, re, glob
import pandas as pd

CSV_IN = "/content/ai-dementia-mri/data/index_oasis1_for3d.csv"
DATA_DIR = "/content/data/processed/oasis1_nifti"
CSV_OUT = "/content/ai-dementia-mri/data/index_oasis1_for3d_FIXED.csv"

assert os.path.exists(CSV_IN), f"Нема CSV: {CSV_IN}"
assert os.path.isdir(DATA_DIR), f"Нема папки з NIfTI: {DATA_DIR}"

# 1) Читаємо початковий індекс і дістаємо meta (subject_id, label, split)
df_raw = pd.read_csv(CSV_IN)
# Узгоджуємо назви колонок
if "nifti_path" not in df_raw.columns and "path" in df_raw.columns:
    df_raw = df_raw.rename(columns={"path":"nifti_path"})

need = {"subject_id","label","split"}
missing = need - set(df_raw.columns)
if missing:
    raise ValueError(f"В індексі бракує колонок {missing}. Є: {list(df_raw.columns)}")

meta = df_raw[["subject_id","label","split"]].drop_duplicates()
print("Meta splits:\n", meta["split"].value_counts(dropna=False), "\n")

# 2) Скануємо фактичні файли в /content/data/processed/oasis1_nifti
files = sorted(glob.glob(os.path.join(DATA_DIR, "*.nii*")))
assert len(files) > 0, f"Не знайшов файлів у {DATA_DIR}"

def extract_sid(p):
    # у файлах типу OAS1_0085_MR1_T1w.nii.gz → subject_id = OAS1_0085_MR1
    m = re.search(r"(OAS1_\d+_MR\d+)", os.path.basename(p))
    return m.group(1) if m else None

paths = pd.DataFrame({"nifti_path": files})
paths["subject_id"] = paths["nifti_path"].map(extract_sid)
paths = paths.dropna(subset=["subject_id"]).drop_duplicates("subject_id")

print("Знайдено NIfTI-файлів (унікальних subject_id):", len(paths))

# 3) Мерджимо мета-дані з фактичними шляхами
df = meta.merge(paths, on="subject_id", how="left")
missing_paths = df["nifti_path"].isna().sum()
print("Записів без знайденого файлу:", missing_paths)

# 4) Фінальний sanity check
print("Розподіл split після злиття:\n", df["split"].value_counts(dropna=False))
print("Розподіл label після злиття:\n", df["label"].value_counts(dropna=False))

# 5) Зберігаємо FIXED CSV
df.to_csv(CSV_OUT, index=False)
print("Готово! Збережено:", CSV_OUT)
```

```python
# src/models/train_3d.py
from __future__ import annotations
import argparse, json, os, math, warnings, random
from pathlib import Path
from typing import Literal, Tuple, Optional, Dict

import numpy as np
import pandas as pd
from tqdm import tqdm

import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import DataLoader, WeightedRandomSampler

from torchvision.models.video import (
    r3d_18, R3D_18_Weights,
    mc3_18, MC3_18_Weights,
    r2plus1d_18, R2Plus1D_18_Weights,
)

from src.data_.dataset3d import VolumeDataset

warnings.filterwarnings("ignore", category=UserWarning)

# -------------------------------
# Утиліти
# -------------------------------
def set_seed(seed: int = 42):
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)
    # детермінізм (може трохи сповільнити)
    torch.backends.cudnn.deterministic = True
    torch.backends.cudnn.benchmark = False

def get_device():
    if torch.cuda.is_available():
        return torch.device("cuda")
    if torch.backends.mps.is_available():
        return torch.device("mps")
    return torch.device("cpu")

def convert_bn_to_gn(module: nn.Module, num_groups: int = 8):
    for name, child in module.named_children():
        if isinstance(child, (nn.BatchNorm1d, nn.BatchNorm2d, nn.BatchNorm3d)):
            ch = child.num_features
            setattr(module, name, nn.GroupNorm(num_groups=min(num_groups, ch), num_channels=ch, affine=True))
        else:
            convert_bn_to_gn(child, num_groups)
    return module

# -------------------------------
# Модель
# -------------------------------
class VideoHead(nn.Module):
    def __init__(self, in_features: int, p: float = 0.2):
        super().__init__()
        self.net = nn.Sequential(
            nn.Dropout(p),
            nn.Linear(in_features, 256),
            nn.ReLU(inplace=True),
            nn.Dropout(p),
            nn.Linear(256, 1)
        )
    def forward(self, x):
        return self.net(x)  # (B,1)

def build_backbone(name: str) -> Tuple[nn.Module, object, int]:
    name = name.lower()
    if name == "r3d_18":
        weights = R3D_18_Weights.KINETICS400_V1
        backbone = r3d_18(weights=weights)
    elif name == "mc3_18":
        weights = MC3_18_Weights.KINETICS400_V1
        backbone = mc3_18(weights=weights)
    elif name == "r2plus1d_18":
        weights = R2Plus1D_18_Weights.KINETICS400_V1
        backbone = r2plus1d_18(weights=weights)
    else:
        raise ValueError(f"Unknown backbone: {name}")
    in_feats = backbone.fc.in_features
    backbone.fc = nn.Identity()
    return backbone, weights, in_feats

class Net3D(nn.Module):
    def __init__(self, backbone_name: str, groupnorm: Optional[int] = None, dropout: float = 0.2):
        super().__init__()
        backbone, weights, in_feats = build_backbone(backbone_name)
        head = VideoHead(in_feats, p=dropout)
        model = nn.Sequential(backbone, head)
        if groupnorm and groupnorm > 0:
            convert_bn_to_gn(model, groupnorm)
            print(f"✅ Converted BatchNorm3d → GroupNorm({groupnorm})")
        self.model = model
        self.weights = weights  # для Kinetics-нормалізації
    def forward(self, x):
        return self.model(x)  # (B,1)

# -------------------------------
# Лосси
# -------------------------------
class FocalLoss(nn.Module):
    def __init__(self, alpha_pos: float = 0.5, gamma: float = 2.0, reduction: str = "mean"):
        super().__init__()
        self.alpha_pos = alpha_pos
        self.gamma = gamma
        self.reduction = reduction
    def forward(self, logits, targets):
        # logits: (B,), targets: (B,)
        p = torch.sigmoid(logits)
        ce = F.binary_cross_entropy_with_logits(logits, targets, reduction="none")
        pt = torch.where(targets == 1, p, 1 - p)
        alpha = torch.where(targets == 1, self.alpha_pos, 1 - self.alpha_pos)
        loss = alpha * (1 - pt).pow(self.gamma) * ce
        if self.reduction == "mean":
            return loss.mean()
        elif self.reduction == "sum":
            return loss.sum()
        return loss

# -------------------------------
# Метрики та агрегація
# -------------------------------
def _aggregate_probs(p: np.ndarray, how: str, topk: int) -> float:
    if how == "mean":
        return float(np.mean(p))
    if how == "median":
        return float(np.median(p))
    if how == "topk":
        k = max(1, min(topk, len(p)))
        return float(np.mean(np.sort(p)[-k:]))
    raise ValueError(how)

def patient_metrics(
    y_true: np.ndarray, y_prob: np.ndarray, sids: list[str],
    thr: float = 0.5, agg: Literal["mean","median","topk"] = "mean", topk: int = 3
) -> Dict[str, float]:
    df = pd.DataFrame(dict(sid=sids, y=y_true, p=y_prob))
    df["y"] = df["y"].astype(int)
    rows = []
    for sid, sub in df.groupby("sid"):
        y = int(round(sub["y"].mean()))
        p = _aggregate_probs(sub["p"].values, agg, topk)
        rows.append((sid, y, p))
    agg_df = pd.DataFrame(rows, columns=["sid","y","p"])
    from sklearn.metrics import roc_auc_score, f1_score, balanced_accuracy_score
    out = {}
    try:
        out["roc_auc"] = float(roc_auc_score(agg_df["y"], agg_df["p"]))
    except Exception:
        out["roc_auc"] = float("nan")
    yhat = (agg_df["p"].values >= thr).astype(int)
    out["f1"] = float(f1_score(agg_df["y"], yhat))
    out["bal_acc"] = float(balanced_accuracy_score(agg_df["y"], yhat))
    return out

# -------------------------------
# Епоха (train/val)
# -------------------------------
def maybe_preprocess_batch(
    x: torch.Tensor,
    device: torch.device,
    kinetics_norm: bool,
    weights_norm: Optional[object],
    zscore: bool,
    clip_lo: Optional[float],
    clip_hi: Optional[float],
):
    # Канали: якщо (B,1,D,H,W) — дублюємо до 3
    if x.size(1) == 1:
        x = x.repeat(1, 3, 1, 1, 1)

    # перцентильний кліпінг у межах тензора batch-wise
    if (clip_lo is not None) or (clip_hi is not None):
        lo = clip_lo if clip_lo is not None else 0.0
        hi = clip_hi if clip_hi is not None else 100.0
        lo_v = torch.quantile(x.reshape(x.size(0), -1), lo/100.0, dim=1, keepdim=True).view(-1,1,1,1,1)
        hi_v = torch.quantile(x.reshape(x.size(0), -1), hi/100.0, dim=1, keepdim=True).view(-1,1,1,1,1)
        x = torch.clamp(x, min=lo_v, max=hi_v)

    # z-score по кожному об'єму (канали окремо)
    if zscore:
        mean = x.mean(dim=(2,3,4), keepdim=True)
        std  = x.std(dim=(2,3,4), keepdim=True).clamp_min(1e-6)
        x = (x - mean) / std

    # Kinetics-нормалізація (очікувана бекбоном)
    if kinetics_norm and (weights_norm is not None):
        mean = torch.as_tensor(weights_norm.mean, device=device)[None, :, None, None, None]
        std  = torch.as_tensor(weights_norm.std,  device=device)[None, :, None, None, None]
        x = (x - mean) / std

    return x

def run_epoch(
    model: nn.Module,
    loader: DataLoader,
    device: torch.device,
    loss_fn: nn.Module,
    optimizer: Optional[torch.optim.Optimizer],
    kinetics_norm: bool,
    weights_norm: Optional[object],
    zscore: bool,
    clip_lo: Optional[float],
    clip_hi: Optional[float],
    train: bool,
    accum_steps: int,
    use_amp: bool,
    clip_grad: Optional[float],
    agg: str,
    topk: int,
):
    model.train(train)
    total_loss, total_n = 0.0, 0
    ys, ps, sids = [], [], []

    scaler = torch.cuda.amp.GradScaler(enabled=(use_amp and device.type == "cuda"))

    for step, (x, y, sid) in enumerate(tqdm(loader, leave=False)):
        x = x.to(device, non_blocking=False)
        y = y.to(device, non_blocking=False).float().view(-1)

        x = maybe_preprocess_batch(
            x, device,
            kinetics_norm=kinetics_norm, weights_norm=weights_norm,
            zscore=zscore, clip_lo=clip_lo, clip_hi=clip_hi
        )

        if train:
            with torch.cuda.amp.autocast(enabled=(use_amp and device.type == "cuda")):
                logits = model(x).view(-1)  # (B,)
                loss = loss_fn(logits, y)   # BCE/Focal
            if scaler.is_enabled():
                scaler.scale(loss/accum_steps).backward()
            else:
                (loss/accum_steps).backward()

            if (step + 1) % accum_steps == 0:
                if clip_grad is not None and clip_grad > 0:
                    if scaler.is_enabled():
                        scaler.unscale_(optimizer)
                    torch.nn.utils.clip_grad_norm_(model.parameters(), clip_grad)
                if scaler.is_enabled():
                    scaler.step(optimizer)
                    scaler.update()
                else:
                    optimizer.step()
                optimizer.zero_grad(set_to_none=True)
        else:
            logits = model(x).view(-1)
            loss = loss_fn(logits, y)

        p = torch.sigmoid(logits).detach().cpu().numpy()
        ys.extend(y.detach().cpu().numpy().tolist())
        ps.extend(p.tolist())
        sids.extend(list(sid))

        total_loss += float(loss.item()) * x.size(0)
        total_n    += x.size(0)

    # slice/volume-level метрики (насправді тут 1 об’єм = 1 «slice»)
    from sklearn.metrics import roc_auc_score, f1_score, balanced_accuracy_score
    ys_np = np.asarray(ys).astype(int)
    ps_np = np.asarray(ps)
    out_slice = {}
    try:
        out_slice["roc_auc"] = float(roc_auc_score(ys_np, ps_np))
    except Exception:
        out_slice["roc_auc"] = float("nan")
    yhat = (ps_np >= 0.5).astype(int)
    out_slice["f1"] = float(f1_score(ys_np, yhat))
    out_slice["bal_acc"] = float(balanced_accuracy_score(ys_np, yhat))

    # підбір порогу по пацієнту — F1
    thrs = np.linspace(0.05, 0.95, 19)
    f1s = []
    for t in thrs:
        f1s.append(patient_metrics(ys_np, ps_np, sids, thr=float(t), agg=agg, topk=topk)["f1"])
    best_thr = float(thrs[int(np.argmax(f1s))])
    out_pat = patient_metrics(ys_np, ps_np, sids, thr=best_thr, agg=agg, topk=topk)

    return total_loss/max(1,total_n), out_slice, out_pat, best_thr

# -------------------------------
# CSV нормалізація + авто-спліт
# -------------------------------
def normalize_and_split_csv(df: pd.DataFrame, data_root_default: str = "/content/data/processed/oasis1_nifti") -> pd.DataFrame:
    df.columns = [str(c).strip().lower() for c in df.columns]

    if "nifti_path" not in df.columns and "path" in df.columns:
        df = df.rename(columns={"path": "nifti_path"})

    if "label" not in df.columns and "cdr" in df.columns:
        cdr_str = df["cdr"].astype(str).str.strip()
        df["label"] = np.where(cdr_str.str.fullmatch(r"(0|0\.0)"), "non_demented", "dementia")

    if "subject_id" not in df.columns:
        def _sid_from_path(p):
            name = Path(str(p)).name
            stem = Path(str(p)).stem
            parts = stem.split("_")
            if len(parts) >= 2 and parts[0].upper().startswith("OAS1"):
                return f"{parts[0]}_{parts[1]}"
            return stem
        df["subject_id"] = df.get("nifti_path", pd.Series(["unknown"]*len(df))).apply(_sid_from_path)

    df["label"] = df["label"].astype(str).str.strip().str.lower()
    df["label"] = df["label"].replace({
        "non-demented": "non_demented",
        "nondemented": "non_demented",
        "control": "non_demented",
        "demented": "dementia"
    })

    need_split = ("split" not in df.columns) or df["split"].isna().all() or (df["split"].astype(str).str.strip() == "").all()
    if need_split:
        print("⚠️ 'split' відсутній/порожній — створюю 70/15/15 по subject_id (seed=42).")
        sids = df["subject_id"].astype(str).dropna().unique()
        rng = np.random.default_rng(42)
        rng.shuffle(sids)
        n = len(sids); n_train = int(0.7*n); n_val = int(0.15*n)
        split_map = {sid: ("train" if i < n_train else "val" if i < n_train + n_val else "test")
                     for i, sid in enumerate(sids)}
        df["split"] = df["subject_id"].map(split_map)
    else:
        df["split"] = df["split"].astype(str).str.strip().str.lower()
        df.loc[df["split"].isin(["valid","validation"]), "split"] = "val"

    base_dir = os.environ.get("DATA_ROOT_3D", data_root_default)
    df["nifti_path"] = df["nifti_path"].astype(str)

    exists_mask = df["nifti_path"].apply(os.path.exists)
    if not exists_mask.all():
        df.loc[~exists_mask, "nifti_path"] = df.loc[~exists_mask, "nifti_path"].apply(
            lambda p: str(Path(base_dir) / Path(p).name)
        )

    df = df[df["nifti_path"].apply(os.path.exists)].reset_index(drop=True)

    need_cols = {"subject_id", "nifti_path", "label", "split"}
    missing = need_cols - set(df.columns)
    assert not missing, f"В індексі бракує колонок: {missing}. Є: {list(df.columns)}"

    print("split counts:\n", df["split"].value_counts(dropna=False))
    return df

# -------------------------------
# Головний цикл
# -------------------------------
def main(args):
    set_seed(args.seed)
    device = get_device()
    print("Device:", device.type)

    df = pd.read_csv(args.index_csv)
    df = normalize_and_split_csv(df, data_root_default="/content/data/processed/oasis1_nifti")

    tr_df = df[df["split"] == "train"].reset_index(drop=True)
    va_df = df[df["split"] == "val"].reset_index(drop=True)

    print("ℹ️ train counts:", dict(tr_df["label"].value_counts()))
    print("ℹ️ val counts:",   dict(va_df["label"].value_counts()))

    # Датасети
    tr_ds = VolumeDataset(tr_df, size=args.size, depth=args.depth, augment=True,  cache=args.cache)
    va_ds = VolumeDataset(va_df, size=args.size, depth=args.depth, augment=False, cache=args.cache)

    # Балансований семплер (за бажанням)
    y_tr = (tr_df["label"].str.lower() == "dementia").astype(int).values
    if args.balanced:
        class_count = np.bincount(y_tr, minlength=2)
        weight = 1.0 / np.maximum(class_count, 1)
        samples_weight = weight[y_tr]
        sampler = WeightedRandomSampler(samples_weight, num_samples=len(samples_weight), replacement=True)
        tr_loader = DataLoader(tr_ds, batch_size=args.batch, sampler=sampler,
                               num_workers=args.num_workers, pin_memory=(device.type=="cuda"))
    else:
        tr_loader = DataLoader(tr_ds, batch_size=args.batch, shuffle=True,
                               num_workers=args.num_workers, pin_memory=(device.type=="cuda"))

    va_loader = DataLoader(va_ds, batch_size=max(1, args.batch*2), shuffle=False,
                           num_workers=args.num_workers, pin_memory=(device.type=="cuda"))

    # Модель
    model = Net3D(args.backbone, groupnorm=args.groupnorm, dropout=0.2)
    model.to(device)

    # Нормалізація під Kinetics
    weights_norm = model.weights.transforms() if args.kinetics_norm else None

    # Лосс
    if args.loss == "bce":
        pos = int((tr_df["label"].str.lower()=="dementia").sum())
        neg = int((tr_df["label"].str.lower()=="non_demented").sum())
        pos_weight = torch.tensor([neg/max(pos,1)], device=device, dtype=torch.float32)
        print(f"pos_weight={float(pos_weight.cpu().numpy()[0]):.3f}")
        loss_fn = nn.BCEWithLogitsLoss(pos_weight=pos_weight)
    else:
        print(f"Focal: alpha_pos={args.alpha:.3f}, gamma={args.gamma}")
        loss_fn = FocalLoss(alpha_pos=args.alpha, gamma=args.gamma)

    # Оптимізатор та шедулери
    opt = torch.optim.AdamW(model.parameters(), lr=args.lr, weight_decay=args.weight_decay)

    # Warmup → Cosine
    def lr_lambda(epoch):
        if epoch < args.warmup_epochs:
            return (epoch + 1) / max(1, args.warmup_epochs)
        # далі — косинус до eta_min = 0.1*lr
        t = epoch - args.warmup_epochs
        T = max(1, args.epochs - args.warmup_epochs)
        return 0.1 + 0.9 * (1 + math.cos(math.pi * t / T)) / 2.0
    sched = torch.optim.lr_scheduler.LambdaLR(opt, lr_lambda=lr_lambda)

    # Заморозка бекбону
    def freeze_backbone(flag: bool):
        for n, p in model.model[0].named_parameters():
            p.requires_grad = not flag

    if args.freeze_epochs > 0:
        freeze_backbone(True)
        print(f"🧊 Freezing backbone for {args.freeze_epochs} epoch(s)")

    best_score = -float("inf")
    best_thr = 0.5
    stale = 0

    for epoch in range(1, args.epochs + 1):
        if epoch == args.freeze_epochs + 1 and args.freeze_epochs > 0:
            print("🔓 Unfreezing backbone")
            freeze_backbone(False)
            # зменшуємо LR при розморозці
            for g in opt.param_groups:
                g["lr"] = args.lr * args.unfreeze_lr_mult

        tr_loss, tr_m_slice, tr_m_pat, tr_thr = run_epoch(
            model, tr_loader, device, loss_fn, opt,
            kinetics_norm=args.kinetics_norm, weights_norm=weights_norm,
            zscore=args.zscore, clip_lo=args.clip_lo, clip_hi=args.clip_hi,
            train=True, accum_steps=args.accum_steps, use_amp=(device.type=="cuda"),
            clip_grad=args.clip_grad, agg=args.agg, topk=args.topk,
        )
        with torch.no_grad():
            va_loss, va_m_slice, va_m_pat, va_thr = run_epoch(
                model, va_loader, device, loss_fn, optimizer=None,
                kinetics_norm=args.kinetics_norm, weights_norm=weights_norm,
                zscore=args.zscore, clip_lo=args.clip_lo, clip_hi=args.clip_hi,
                train=False, accum_steps=1, use_amp=False,
                clip_grad=None, agg=args.agg, topk=args.topk,
            )

        print(
            f"Epoch {epoch:02d} | train_loss={tr_loss:.4f} | val_loss={va_loss:.4f} | "
            f"slice: AUC={tr_m_slice['roc_auc']:.3f}->{va_m_slice['roc_auc']:.3f} "
            f"F1={va_m_slice['f1']:.3f} BA={va_m_slice['bal_acc']:.3f} | "
            f"patient@thr={va_thr:.2f}: AUC={va_m_pat['roc_auc']:.3f} "
            f"F1={va_m_pat['f1']:.3f} BA={va_m_pat['bal_acc']:.3f} | lr={opt.param_groups[0]['lr']:.2e}"
        )

        # Вибір метрики для ранньої зупинки
        cur = va_m_pat["roc_auc"] if args.monitor == "pat_auc" else va_m_pat["bal_acc"]

        # Збереження чекпойнтів
        Path("checkpoints").mkdir(exist_ok=True, parents=True)
        torch.save(model.state_dict(), "checkpoints/last_3d.pt")
        improved = (cur > best_score + args.min_delta)
        if improved:
            best_score = cur
            best_thr = va_thr
            stale = 0
            torch.save(model.state_dict(), "checkpoints/best_3d.pt")
            Path("artifacts").mkdir(exist_ok=True, parents=True)
            with open("artifacts/thresholds_3d.json", "w") as f:
                json.dump({"metric": args.monitor, "thr": best_thr, "agg": args.agg, "topk": args.topk}, f, indent=2)
            print(f"  ↳ ✅ saved: checkpoints/best_3d.pt (score={best_score:.3f}, thr={best_thr:.2f})")
        else:
            stale += 1
            if stale >= args.patience:
                print(f"⏹ Early stopping ({args.patience}), best {args.monitor}={best_score:.3f} @ thr={best_thr:.2f}")
                break

        sched.step()

    print(f"Best {args.monitor}: {best_score:.3f} @ thr={best_thr:.2f}")

# -------------------------------
# CLI
# -------------------------------
def parse_args():
    ap = argparse.ArgumentParser()
    ap.add_argument("--index_csv", type=str, required=True)

    # тренування
    ap.add_argument("--epochs", type=int, default=50)
    ap.add_argument("--warmup_epochs", type=int, default=0)
    ap.add_argument("--batch", type=int, default=2)
    ap.add_argument("--size", type=int, default=128)
    ap.add_argument("--depth", type=int, default=64)

    # оптимізація
    ap.add_argument("--lr", type=float, default=1e-4)
    ap.add_argument("--weight_decay", type=float, default=5e-4)

    # фріз/анфріз
    ap.add_argument("--freeze_epochs", type=int, default=4)
    ap.add_argument("--unfreeze_lr_mult", type=float, default=0.3)

    # рання зупинка
    ap.add_argument("--patience", type=int, default=8)
    ap.add_argument("--min_delta", type=float, default=0.0)
    ap.add_argument("--monitor", type=str, choices=["pat_auc","pat_ba"], default="pat_auc")

    # даталоадер
    ap.add_argument("--num_workers", type=int, default=0)
    ap.add_argument("--accum_steps", type=int, default=4)

    # нормалізація
    ap.add_argument("--groupnorm", type=int, default=8)
    ap.add_argument("--balanced", action="store_true")
    ap.add_argument("--cache", action="store_true")
    ap.add_argument("--zscore", action="store_true")
    ap.add_argument("--clip_lo", type=float, default=None, help="перцентиль нижній (напр. 0.5)")
    ap.add_argument("--clip_hi", type=float, default=None, help="перцентиль верхній (напр. 99.5)")
    ap.add_argument("--kinetics_norm", action="store_true", default=True)
    ap.add_argument("--backbone", type=str, default="r3d_18", choices=["r3d_18","mc3_18","r2plus1d_18"])

    # агрегація пацієнта
    ap.add_argument("--agg", type=str, choices=["mean","median","topk"], default="mean")
    ap.add_argument("--topk", type=int, default=3)

    # лосс
    ap.add_argument("--loss", type=str, choices=["bce","focal"], default="bce")
    ap.add_argument("--gamma", type=float, default=2.0)
    ap.add_argument("--alpha", type=float, default=0.6)

    # стабільність
    ap.add_argument("--clip_grad", type=float, default=None)
    ap.add_argument("--seed", type=int, default=42)

    return ap.parse_args()

if __name__ == "__main__":
    args = parse_args()
    os.makedirs("artifacts", exist_ok=True)
    main(args)
```
